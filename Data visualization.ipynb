```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('cancer_data_set_batch_3.csv')
df
```

```python
df.shape
```

```python
df.columns
```

```python
print(df.isna().sum())
```

```python
#Dropping the Unnamed column
df = df.drop(["Unnamed: 32"],axis=1)
df
```

```python
#Removoving the null values 
df = df.dropna(axis=1)
df
```

```python
#Encoding categorical data 
#Making Male as 1 and Female as 0
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

# Encode labels in column 'diagnosis'.
df['diagnosis']= le.fit_transform(df['diagnosis'])
  
df['diagnosis'].unique()
df
```

```python
df.describe()
```

```python
df = df.drop('id',axis=1)
```

```python
df.columns
```

```python
# *********************************** 1.Scatter Plot **************************************
```

```python
'''A scatter plot of *y* vs. *x* with varying marker size and/or color.'''
```

```python
# x --> x data 
# y --> y data 
# color --> color
```

```python
plt.scatter(df['radius_mean'],df['texture_mean'],color = "GREEN")
plt.xlabel('Radius mean')
plt.ylabel('Texture mean')
plt.show()
```

```python
# ******************************************** 2.BAR Plot *******************************
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x="diagnosis", data=df, palette="pastel")
plt.title("Diagnosis Count")
plt.show()
```

```python
# ********************************** 3.PIE Plot ***********************************
```

```python
#autopct --> percentage 
#explode --> add gap between the entites (3D View)
#startangle --> starting angle 
#title --> title for graph
```

```python
plt.figure(figsize=(10, 6))
df["diagnosis"].value_counts().plot(kind="pie", autopct="%1.1f%%", explode=[0,0.1], 
                                     colors=["red", "green"], startangle=90)
plt.title("Diagnosis Proportion")
plt.ylabel("")
plt.show()
```

```python
# ******************************* 4.Area Plot ***************************************
```

```python
'''Kdeplot allows us to estimate the probability density function of the continuous or non-parametric 
from our data set curve in one or more dimensions it means we can create plot a single graph 
For multiple samples which helps in more efficient data visualization.'''
```

```python
# data --> dataframe 
# x --> X values 
# hue --> data to visualize 
# fill --> making the plot solid by filling the colours 
# alpha --> Intensity of the colors 
# linewidth --> width of the line 
```

```python
plt.figure(figsize=(12, 8))
sns.kdeplot(data=df, x="radius_mean", hue="diagnosis", fill=True, alpha=0.5, 
            linewidth=0, palette="bright")
plt.title("Radius Mean Distribution")
plt.xlabel("Radius Mean")
plt.ylabel("Density")
plt.show()
```

```python
# *********************************** 5.Histogram Plot *******************************
```

```python
# ***************************************** 10.Regression Plot *****************************
```

```python
'''Plot data and a linear regression model fit.'''
```

```python
# x --> x data 
# y --> y data 
# data --> datafram
# scatter_kws --> scatter point properties 
# line_kws --> line properties 
```

```python
# ********************************** 21. Heatmap with clustering *******************************
```

```python
# corr ->> Colrealtions 
# annot --> visibility of corelations 
# row_cluster ---> Enable row cluster 
# Col_cluster --> Enable col clusetr 
```

```python
corr = df.corr()
sns.clustermap(corr, cmap='coolwarm', annot=True, linewidths=.5, figsize=(10, 10), row_cluster=True, col_cluster=True, cbar_pos=(0, 0.2, 0.03, 0.4), dendrogram_ratio=0.1)
```
