```python
#Dropping the Unnamed column
df = df.drop(["Unnamed: 32"],axis=1)
```

```python
#Removoving the null values 
df = df.dropna(axis=1)
df
```

```python
#Encoding categorical data 
#Making Male as 1 and Female as 0
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

# Encode labels in column 'diagnosis'.
df['diagnosis']= le.fit_transform(df['diagnosis'])
  
df['diagnosis'].unique()
df
```

```python
# Data mearging 
# It overrides the perimeter_mean 

data = pd.merge(data1,data2)
data
```

```python
new_row = {'id':670, 'diagnosis':1, 'radius_mean':20, 'texture_mean':50, 'perimeter_mean':50,
       'area_mean':50, 'smoothness_mean':90, 'compactness_mean':90, 'concavity_mean':90,
       'concave points_mean':90, 'symmetry_mean':90, 'fractal_dimension_mean':90,
       'radius_se':90, 'texture_se':90, 'perimeter_se':90, 'area_se':90, 'smoothness_se':90,
       'compactness_se':90, 'concavity_se':90, 'concave points_se':90, 'symmetry_se':90,
       'fractal_dimension_se':90, 'radius_worst':90, 'texture_worst':90,
       'perimeter_worst':90, 'area_worst':90, 'smoothness_worst':90,
       'compactness_worst':90, 'concavity_worst':90, 'concave points_worst':90,
       'symmetry_worst':90, 'fractal_dimension_worst':90}
```

```python
# ****************************** 5. Model Building ********************************
```

```python
from sklearn import linear_model
```

```python
#Defining feature set
X = data.iloc[:,2:]
X
```

```python
#Label
y = data.iloc[:,1:2]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
#Using tree classifier for model Building 
from sklearn import tree
tree_clas = tree.DecisionTreeClassifier(criterion='entropy')
tree_clas.fit(X_train,y_train)
```

```python
y_test_predict = tree_clas.predict(X_test)
y_test_predict
```

```python
np.array(y_test)
```

```python
#Prediction on training dataset
y_train_predict = tree_clas.predict(X_train)
y_train_predict
```

```python
np.array(y_train)
```

```python
#Calculating the accuracy of prediction on the training set
from sklearn.metrics import accuracy_score

test_accuracy = accuracy_score(y_test,y_test_predict)
print(test_accuracy)
```

```python
train_accuracy = accuracy_score(y_train,y_train_predict)
print(train_accuracy)
#this model is overfitted due to accuracy is 1 
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_test_predict)
```
